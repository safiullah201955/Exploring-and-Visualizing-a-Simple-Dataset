# **Problem Statement**
## Perform exploratory data analysis (EDA) on the Iris dataset to explore features, visualize distributions, and prepare insights for building future ML models for classification.

```python
#Import Required Libraries

import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load the Dataset
df = sns.load_dataset("iris")
```

```python
#Pairwise Relationships

sns.pairplot(data = df,hue="species")
plt.suptitle("Pairplot of Iris Dataset", fontsize=16, y=1.02)
plt.show()
```

```python
#Distribution Plots for Each Feature

features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

for feature in features:
    plt.figure(figsize=(6, 4))
    sns.histplot(data=df, x=feature)
    plt.title(f"Distribution of {feature}")
    plt.xlabel(feature)
    plt.ylabel("Count")
    plt.show()
```

```python
#Box Plots for Each Feature by Species

features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]

for feature in features:
    plt.figure(figsize=(6, 4))
    sns.boxplot(data=df)
    plt.title(f"Box Plot of {feature} by Species")
    plt.show()
```
